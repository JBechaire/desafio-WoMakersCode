DESAFIO 1

Como analista de dados em uma empresa de tecnologia, você precisa criar uma função para calcular uma métrica simples de diversidade de gênero na liderança. A função deve receber o número total de líderes e o número de líderes mulheres.

Calcule a porcentagem de mulheres na liderança e retorne uma classificação:

* **Abaixo de 30%:** "A empresa está abaixo da meta de diversidade."
* **De 30% a 50%:** "A empresa está na meta de diversidade."
* **Acima de 50%:** "A empresa excedeu a meta de diversidade."

```python
print("=== Cálculo de Diversidade na Liderança ===")

# O usuário informa o total de líderes
total_lideres = int(input("Digite o TOTAL de líderes da empresa: "))

# O usuário informa quantas líderes são mulheres
lideres_mulheres = int(input("Digite o número de líderes MULHERES: "))

# Calculando o percentual de mulheres
percentual = (lideres_mulheres / total_lideres) * 100

# Mostrando o valor calculado
print(f"\nPercentual de mulheres na liderança: {round(percentual, 1)}%")

# critérios 
if percentual < 30:
    print("A empresa está abaixo da meta de diversidade.")
elif percentual <= 50:
    print("A empresa está na meta de diversidade.")
else:
    print("A empresa excedeu a meta de diversidade.")
```

Para acessar a plataforma do bootcamp, as participantes precisam realizar um login simples.

Crie um programa que simule esse processo utilizando um usuário e senha pré-definidos.
O programa deve solicitar os dados e informar se o acesso foi permitido ou negado.

```python
print("=== Sistema de Login Simples ===")
# usuário e senha corretos (pré-definidos)
usuario_correto = "WoMakers"
senha_correta = "Code@123"

# A pessoa digita o usuário e a senha
# A função input lê o que o usuário digitar no teclado como texto (string)
usuario_digitado = input("Digite o usuário: ")
senha_digitada = input("Digite a senha: ")

# Compara o que foi digitado com o que é correto
# '==' para verificar se os textos são exatamente iguais
if usuario_digitado == usuario_correto and senha_digitada == senha_correta:
    # Se as duas comparações forem verdadeiras, o acesso é permitido
    print("Acesso permitido.")
else:
    # Caso contrário (qualquer um estiver errado), acesso negado
    print("Acesso negado.")
```

DESAFIO 3

Você recebeu uma lista de dicionários com dados de funcionárias de uma empresa de tecnologia. Sua tarefa é calcular o salário médio das mulheres que trabalham no departamento de "Engenharia de Dados".

Dica: Você precisará percorrer a lista, verificar se a funcionária é do departamento correto e, em caso afirmativo, somar seu salário e contar quantas são para depois calcular a média.




```python
funcionarias = [
    {"nome": "Beatriz", "departamento": "Análise de BI", "salario": 7000},
    {"nome": "Helena", "departamento": "Engenharia de Dados", "salario": 9500},
    {"nome": "Isabela", "departamento": "Ciência de Dados", "salario": 11000},
    {"nome": "Laura", "departamento": "Engenharia de Dados", "salario": 8500},
    {"nome": "Mariana", "departamento": "Engenharia de Software", "salario": 7500}
]
```

```python
print("===  Média Salarial no Departamento de Engenharia de Dados ===")

# Departamento que queremos filtrar
departamento = "Engenharia de Dados"

#somar salários e contar quantas funcionárias são do departamento
soma_salarios = 0
quantidade = 0

# Percorre (loop) cada funcionária da lista
for func in funcionarias:
    # Verifica se o departamento da funcionária é o que queremos
    if func["departamento"] == departamento:
        # Se for, somamos o salário dela
        soma_salarios = soma_salarios + func["salario"]
    
        quantidade = quantidade + 1

# calcula e imprime a média se encontrou pelo menos 1 funcionária
if quantidade > 0:
    media = soma_salarios / quantidade
   
    print(f"Média salarial em '{departamento}': R$ {round(media, 2)}")
else:
    print(f"Não há funcionárias no departamento '{departamento}'.")
```

DESAFIO 4

Durante a consolidação de dados, foi identificada a presença de valores repetidos em uma lista numérica.

Crie um programa que receba uma lista de números e gere uma nova lista sem valores duplicados, mantendo apenas um valor de cada número.

```python
print("=== Remover Valores Duplicados de uma Lista ===")

lista_numeros = [5, 2, 8, 5, 3, 2, 9]

lista_sem_duplicados = []

# Percorre cada número da lista original
for numero in lista_numeros:
    # Se o número AINDA NÃO está na nova lista, adicionamos
    if numero not in lista_sem_duplicados:
        lista_sem_duplicados.append(numero)

print("Lista original:       ", lista_numeros)
print("Sem valores repetidos:", lista_sem_duplicados)
```

A gerente de projetos, Letícia, precisa de uma ferramenta simples para calcular quantos dias faltam para a entrega de diferentes projetos.

Crie um programa que importe o módulo `datetime`. Peça à usuária para inserir a data de entrega de um projeto no formato **DD/MM/AAAA**. Calcule e exiba o número de dias restantes entre a data de hoje e a data de entrega.

```python
# Importacao datetime 
from datetime import datetime, date

print("=== Dias Restantes para Entrega de Projetos ===")

# Pedir data 
data_str = input("Digite a data de entrega (DD/MM/AAAA): ")

#    Transformamos o texto (string) em uma data de verdade
#    Usamos datetime.strptime para "interpretar" a string conforme o formato indicado
#    Exemplo: "25/12/2026" vira um objeto datetime; depois pegamos só a parte 'date'
data_entrega = datetime.strptime(data_str, "%d/%m/%Y").date()

# Pegamos a data de hoje
hoje = date.today()

#    Calculamos a diferença entre as datas (entrega - hoje)
#    O resultado é um objeto 'timedelta'; para obter os dias inteiros, usamos .days
dias_restantes = (data_entrega - hoje).days

# resultado
if dias_restantes > 0:
    print(f"Faltam {dias_restantes} dia(s) para a entrega.")
elif dias_restantes == 0:
    print("A entrega é hoje!")
else:
    print(f"O prazo já passou há {abs(dias_restantes)} dia(s).")
```
